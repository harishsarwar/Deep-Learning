```python
# we now under the data cleaning process and we noticed that there are so many columns having the data data into the string form lets clean that 
#data and converted them into the integers..
df1.head()
```

```python
def print_unique_col_value(n):
    for columns in n:
        if n[columns].dtype == 'object':
            print(f'{columns} : {n[columns].unique()}')
```

```python
print_unique_col_value(df1)
```

```python
# here in the columns data we can watch there are alots of another formates data like'no internet service and no phone service '
# lets replace this syntex with simple yes or no formates..
df1 = df1.replace('No phone service', 'No')
df1 = df1.replace('No internet service', 'No')
```

```python
# finction called by name ...
print_unique_col_value(df1)
```

```python
df1.columns
```

```python
# converted yes no into the 0_1 formate..
yes_no_columns = ['Partner', 'Dependents', 'PhoneService', 'MultipleLines','OnlineSecurity',
                  'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV','StreamingMovies','PaperlessBilling','Churn']
```

```python
for col in yes_no_columns:
    df1[col].replace({
        'Yes': 1,
        'No' : 0,
    }, inplace= True)

# all set.. all columns are going to converted into the range of 0 and 1..
df1.head()
```

```python
# gender column replace into the range between 0 and 1..
df1.gender.replace({
    'Male':1,
    'Female':0
}, inplace=True)
```

```python
df1.head()
```

```python
#['InternetService', 'Contract', 'PaymentMethod']. these columns contains categorical variables.which contains string..
# lets convert them by using hot encoding...

# HOT encoding...= nothink but convert categoricall string data into the form of intezer......
# HOt encoding by called get_dummies function into the pandas dataframe

df2 = pd.get_dummies(data = df1, columns=['InternetService', 'Contract', 'PaymentMethod'])
df2.head()
```

```python
# df2 contains boolean.

df2 = df2*1  #(converting boolean to integer)
df2.head()
```

```python
df2.dtypes
```

```python
df2.info()
```

```python
# data cleaning work complete all set data fully cleaned up and contains integersss...
```

```python
# ANN(artifitial neural network) algorithm in deep learning apply...
```

```python
# scaling is must important inthe the deeplearning bcz its gives the accuracy to the model..
#scaling is nothing but convert data into the range of (0-1)..

scl_column =['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler # convert data into the range of 0-1
scale = MinMaxScaler()
df2[scl_column] = scale.fit_transform(df2[scl_column])
df2[['tenure','MonthlyCharges','TotalCharges']]
```

```python
df2
```

```python
# train_test_split method used before applying the machine learing algorithm..
from sklearn.model_selection import train_test_split

X = df2.drop('Churn', axis='columns')
y = df2.Churn
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
len(X_train)
```

```python
X_train.shape
```

```python
len(X_test)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    # dont have 2d array so there are need for flatted dense.
# Note:input shape contains shape of training data (contains total number of columns..)
    keras.layers.Dense(50,input_shape=(26,),activation = 'relu'),#.. Hidden layer with 50 hidden neurons
    keras.layers.Dense(1 , activation='sigmoid')  #.. output layer wiht 1 output layer bcz prediction having 2 possibility 0 and 1
])
model.compile(optimizer='adam',
              loss='binary_crossentropy', # binary_crossentropy because our output is binary..
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs = 10)
```

```python
model.evaluate(X_test,y_test)
```

```python
pred = model.predict(X_test)
pred[:5]
```

```python

# convering 2d of predicton into 1d..
predictions = []
for no in pred:
    if no > 0.50:
        predictions.append(1)
    else:
        predictions.append(0)
predictions[:3]
```
