```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
df = pd.read_csv('insurance_data (1).csv')
```

```python
df.head()
```

```python
df.shape
```

```python
# Scaling performed for converting value which is available in range 1 -100 in to the range of 0 -1..
#Scaling is way to increases the accurancy in the model 
```

```python
# scaling converting age into the range of 0-1 becz all values present into the range of 0-1...
#when data set represent into the same range its give more accuracy
df.age= df.age / 100
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['age', 'affordibility']]
```

```python
y = df.bought_insurance
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
len(X_train)
```

```python
X_train
```

```python
# Now using tensorflow module we going to create a simple neural network having not any hidden layers..
```

```python
model = keras.Sequential([
    keras.layers.Dense(1, input_shape = (2,), activation = 'sigmoid',kernel_initializer='ones',bias_initializer='zeros' )
])
'''in model 1 output layere bcz output returns weither the insurance is bought or not, and having input shape 2 represented into the X_train
And other repersentation available inthe the code basics lacture's diagrame....'''
model.compile(
    optimizer='adam',
    loss = 'binary_crossentropy',
    metrics=['accuracy']
)
model.fit(X_train,y_train,epochs=200)
```

```python
model.evaluate(X_test,y_test)
```

```python
prediction = model.predict(X_test)
```
