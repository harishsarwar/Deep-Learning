# Regularization:
used to solve overfitting and underfitting problems.


```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
df =  pd.read_csv('sonar_dataset.csv', header=None)
df.head()
```

```python
df[60]. value_counts()
```

```python
df.dtypes
```

```python
#1.. way to convert object into intezer(using pandas replace method)

# df[60] = df[60].replace({
#     'R' : 1,
#     'M' : 0
# })

# 2...one hot encoding..
df1 = pd.get_dummies(df[60])
df2 = df1['M']
df2
```

```python
df.drop(60,axis='columns',  inplace=True)
df
```

```python
df = pd.concat([df,df2], axis='columns')
df.head()
```

```python
df = df.rename(columns = {'M': 60})
df = df*1
df
```

```python
X = df.drop(60,axis='columns')
y = df[60]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape, y_train.shape
```

```python
# creating ANN 
```

```python
model1 = keras.Sequential({
    keras.layers.Dense(60, input_shape = (60,), activation = 'relu'),
    keras.layers.Dense(30, activation='relu'),  #hidden
    keras.layers.Dense(15, activation = 'relu'), #hidden
    keras.layers.Dense(1, activation = 'sigmoid' )#output 
})
model1.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy'] 
    
)
model1.fit(X_train,y_train, epochs=100)
```

```python
model1.evaluate(X_test,y_test)
```

```python
prediction = model1.predict(X_test).reshape(-1) # direct flettend using numpy module..
prediction[:10]
```

```python
# predition are available  into the whole number form and prediction happend with sigmoid finctoin 
#converting it into the sigmoid fuction..
prediction = np.around(prediction) # round method is a way to convert whole number into the sigmoid (0-1) range..
prediction[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(prediction,y_test))
```

```python
# Drop out Regularization : this technique used for handle the problem of overfitting..in this drop some neurons ramdomely from the hidden layer..
```

```python
#code same but apply regularization technique.

# regularization(keras.layers.Dropout) applies after every hidden layers..

model2 = keras.Sequential([
    keras.layers.Dense(60, input_shape = (60,),activation = 'relu'),
    keras.layers.Dropout(0.5), # (0.5) shows 50 percent of neurons layers gone dropout
    keras.layers.Dense(30, activation = 'relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15,activation = 'relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1,activation='sigmoid')
])
model2.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
model2.fit(X_train,y_train, epochs = 100)
```

```python
# compare here the accuracy of boht model some times it gives equal but some time gives better after apply regularization....
# all set model2 gives better accuracy than model1...

model2.evaluate(X_test,y_test)
```

```python
prediction2 =  model2.predict(X_test)
```

```python
# flattend apply for converting 2d prediction2 data into 1d data.. by using for loop and append list method..
pred = []
for no in prediction2:
    if no > .5 :
        pred.append(1)
    else:
        pred.append(0)

# another way using pandas (reshape method).
        
# prediction2 = model.predict(X_test).reshape(-1)
```

```python
pred[:10]
```

```python
y_test[:5]
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, pred))
```

```python
print(confusion_matrix(y_test, pred))
```
